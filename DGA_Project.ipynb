```python
extracted_domains=df[0].apply(tldextract.extract)
extracted_domains
```

```python
extracted_df=pd.DataFrame(extracted_domains.to_list(),columns=["Subdomain","Domain","Suffix"])
extracted_df
```

```python
extracted_df["Domain_length"]=extracted_df["Domain"].apply(len)
extracted_df
```

```python
extracted_df['Vowels_count'] = extracted_df.Domain.str.lower().str.count(r'[aeiou]')
extracted_df['Consonant_count'] = extracted_df.Domain.str.lower().str.count(r'[a-z]') -extracted_df['Vowels_count']
extracted_df
```

```python
extracted_df['Vowels'] = extracted_df.Domain.str.lower().str.extractall(r'([aeiou])')[0].groupby(level=0).apply(''.join)
extracted_df['Consonant'] = extracted_df.Domain.str.lower().str.extractall(r'([bcdfghjklmnpqrstvwxyz])')[0].groupby(level=0).apply(''.join)
extracted_df
```
